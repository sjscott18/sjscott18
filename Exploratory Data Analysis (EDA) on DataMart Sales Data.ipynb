```python
# Import pandas, numpy, matplotlib and seaborn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set up matplotlib to display plots inline
%matplotlib inline

# Download the "Global Superstore Dataset" from Kaggle and use pandas to read the CSV
df=pd.read_csv('C:/Users/sjsco/Desktop/Datasets/Global_Superstore.csv')

# Display the first few rows of the data using head()
df.head()
```

```python
# Check the shape of the DataFrame
print(df.shape)

# Display info about the DataFrame using info()
df.info()

# Generate descriptive statistics using describe()
df.describe()
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Handle any missing or inconsistent data
df.dropna(inplace=True)
```

```python
# Convert data types if necessary
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])


# Calculate total sales and profit
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()

print(f"Total Sales: {total_sales}, Total Profit: {total_profit}")
```

```python
df.info()
```

```python
# Identify top selling products
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
print(top_products)
```

```python
# Analyse sales by region
sales_by_region=df.groupby('Region')['Sales'].sum()
print(sales_by_region)

# Examine the relationship between sales and profit
correlation = df['Sales'].corr(df['Profit'])
print(f"Correlation between Sales and Profit: {correlation}")
```

```python
# Create a bar chart of sales by product category
sns.barplot(x='Category', y='Sales', data=df)
plt.title('Sales by Product Category')
plt.show()
```

```python
# Plot a line graph of monthly sales over time

# Create a new 'Month' column from 'Order Date'
df['Month'] = df['Order Date'].dt.to_period('M')

# Calculate monthly sales
monthly_sales = df.groupby('Month')['Sales'].sum()

# Ensure the index is in the datetime format for plotting
monthly_sales.index = monthly_sales.index.to_timestamp()


monthly_sales.plot(kind='line')
plt.title('Monthly Sales Over Time')
plt.show()
```

```python
# Generate a scatter plot of sales vs. profit
sns.scatterplot(x='Sales', y='Profit', data=df)
plt.title('Sales vs. Profit')
plt.show()
```

```python
# Produce a heatmap of sales by region and product category

# Pivot table for Sales by Region and Product Category
pivot_table = df.pivot_table(values='Sales', index='Region', columns='Category', aggfunc='sum')

sns.heatmap(pivot_table, annot=True, fmt='.0f', cmap='YlGnBu')
plt.title('Sales by Region and Product Category')
plt.show()
```

```python
# Perform a simple time series analysis to indentify any sales trends or seasoanlity
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(monthly_sales, model='additive', period=12)
result.plot()
plt.show()
```

```python
#total_discount, total_profit, discount_profit_correlation

# Calculate total discount and total profit
total_discount = df['Discount'].sum()
total_profit = df['Profit'].sum()

# Convert 'Discount' and 'Profit' to numeric
df['Discount'] = pd.to_numeric(df['Discount'], errors='coerce')
df['Profit'] = pd.to_numeric(df['Profit'], errors='coerce')

# Calculate the correlation between discount and profit
discount_profit_correlation = df[['Discount', 'Profit']].corr().iloc[0, 1]

print(total_discount)
print(total_profit)
print(discount_profit_correlation)
```
